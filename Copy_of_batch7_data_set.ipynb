Data Distribution:
it is a process of that Distribute the data based on type of distribution

Numpy Random Module:it work out randomly,so it used in distribution

There are three types of distributions
1.Normal Distribution
2.Uniform Distribution
3.Logistic Distribution

````python
#Normal Distribution:
import numpy as np
x=np.random.normal(1,2,10)
print(x)
````

````python
import matplotlib.pyplot as plt
x=np.random.normal(1,2,10)
print(x)
plt.hist(x,100)
plt.show()
````

````python
#Uniform Distribution:
import numpy as np
x=np.random.uniform(1,2,10)
print(x)
````

````python
import matplotlib.pyplot as plt
x=np.random.uniform(1,2,10)
print(x)
plt.hist(x,100)
plt.show()
````

````python
#Logistic distribution:import numpy as np
x=np.random.uniform(1,2,10)
print(x)
````

````python
import matplotlib.pyplot as plt
x=np.random.logistic(1,2,10)
print(x)
plt.hist(x,100)
plt.show()
````
